## To detect signal drop I will check some (e.g. 32) frames before and after 'silent' part:

```python
def drop_begins_suddenly(wave_data, begin_frame_index, num_frames_before = 32, treshold_before = 256):
    fb_count = 0.0
    fb_sum = 0.0
    for i in range(max(begin_frame_index-num_frames_before, 0), begin_frame_index):
        fb_sum += abs(wave_data[i])
        fb_count += 1
    if fb_count > 0:
        fb_mean = fb_sum / fb_count
    else:
        fb_mean = 0
    if fb_mean > treshold_before:
        return True
    else:
        return False
```

```python
def drop_ends_suddenly(wave_data, end_frame_index, num_frames_after = 32, treshold_after = 256):
    fb_count = 0.0
    fb_sum = 0.0
    for i in range(end_frame_index, min(end_frame_index+num_frames_after, wave_data.size)):
        fb_sum += abs(wave_data[i])
        fb_count += 1
    if fb_count > 0:
        fb_mean = fb_sum / fb_count
    else:
        fb_mean = 0
    if fb_mean > treshold_after:
        return True
    else:
        return False
```

```python
def detect_signal_drop(file_name, ls_treshold = 2, ls_min_length = 64, 
                       num_frames_before = 32, num_frames_after = 32, 
                       treshold_before = 164, treshold_after = 164):
    wave_read = wave.open(file_name)
    str_data = wave_read.readframes(wave_read.getnframes())
    wave_read.close()
    wave_data = np.frombuffer(str_data, dtype=np.short)
    
    fn = 0 # frame number
    low_signal = False
    
    ls_start = 0
    ls_length = 0
    
    while fn < wave_data.size:
        if wave_data[fn] < ls_treshold:
            if low_signal:
                ls_length += 1
            else:
                ls_start = fn
                ls_length = 1
                low_signal = True
        elif low_signal:
                if (ls_length > ls_min_length) and drop_begins_suddenly(wave_data, ls_start) and drop_ends_suddenly(wave_data, fn):
                    print (file_name + ": invalid (fr. " + str(ls_start) + "-" + str(fn) + ")")
                    return
                low_signal = False
        fn += 1
        
    print (file_name + ": valid")
```

```python
detect_signal_drop("nan-ai-file-1.wav")
```

```python
detect_signal_drop("nan-ai-file-2.wav")
```

```python
detect_signal_drop("nan-ai-file-3.wav")
```
